## 1. While 문

### 1.1. while 문 만들기

```python
while 조건문:
  실행문
```

```python
hit = 0
while hit < 10:
  hit = hit + 1
  print(f'나무를 {hit}번 찍었습니다.')
  if hit == 10:
    print('나무가 넘어갔습니다!')
```

### 1.2. while문 빠져나가기

```python
coffee = 10
money = 500000
while money:
  money = money - 2500
  print(f'돈을 받았으니 커피를 줍니다!(현금 = {money})')
  coffee = coffee - 1
  print(f'남은 커피의 양은 {coffee}잔 입니다.')
  if coffee == 0:
    print('오늘 커피가 다 떨어졌습니다ㅠㅠ')
    break
```

* 반복 더하기(사칙연산 모두 가능)

```python
number = 0
while number < 5:
  number = number + 1
  print(number)
```

```python
number = 0
while number < 5:
  number += 1
  print(number)
```

* for 문과 if문의 결합

```python
scores = [72, 80, 99, 55, 61, 11, 12, 54]
number = 0
for score in scores:
  number += 1 # 반복 더하기!
  if score >= 70:
    result = '합격'
  else:
    result = '불합격'
  print(f'{number}번째 학생의 점수는 {score}점으로 {result}입니다.')
```

* continue 문의 사용: 뒷문장을 실행하지 않고, 처음으로 돌아간다.

```python
scores = [72, 55, 61, 80, 11, 99, 12, 54]
number = 0
for score in scores:
  number += 1 # 반복 더하기!
  if score < 70:
    continue
  print(f'{number}번째 학생의 점수는 {score}점으로 합격입니다.')
```

### quiz. 
아래 코드를 수정해서, 합격, 불합격이 아니라, 100 \~ 71까지는 A, 70 \~ 51까지는 B, 50점 이하는 C가 나오도록 수정하시오.

* hint: elif

```python
scores = [72, 80, 99, 55, 61, 11, 12, 54]
number = 0
for score in scores:
  number += 1 # 반복 더하기!
  if score > 70:
    result = 'A'
  elif score > 50:
    result = 'B'
  else:
    result = 'C'
  print(f'{number}번째 학생의 점수는 {score}점으로 {result}입니다.')
```

### 2.2. range문 사용

```python
for i in range(10):
  print(i)
```

```python
for i in range(1, 11):
  print(i)
```

```python
for i in range(1, 11, 2):
  print(i)
```

* range(end)
  * range(5) = [0,1,2,3,4]
* range(start, end)
  * range(1,6) = [1,2,3,4,5]
* range(start, end, step)
  * range(1, 6, 2) = [1, 3, 5]

```python
scores = [72, 80, 99, 55, 61, 11, 12, 54]
print(len(scores))
for i in range(len(scores)):
  if scores[i] >= 70:
    result = '합격'
  else:
    result = '불합격'
  print(f'{i+1}번째 학생의 점수는 {scores[i]}점입니다.')
```
